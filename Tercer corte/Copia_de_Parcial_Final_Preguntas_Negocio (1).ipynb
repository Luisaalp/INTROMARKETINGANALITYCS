## 3. Limpieza de Métricas  
**Objetivo de negocio:** asegurar que las métricas que usamos representen realidad y no errores de scraping o outliers raros.

### 3.1. Preguntas de negocio

1. **Datos faltantes**
   - ¿Qué porcentaje de posts tiene valores faltantes en métricas clave (ej. `likes`, `comments`, `views`)?  
   - Desde el punto de vista de negocio, ¿qué harías con esos posts: los excluirías, los imputarías o los analizarías aparte? ¿Por qué?

2. **Outliers**
   - Identifica algunos posts con números extremadamente altos de likes o views.  
   - ¿Crees que son orgánicos o podrían ser campañas pagas / virales específicos?  
   - ¿Cómo cambia tu lectura si incluyes vs. excluyes estos outliers?

3. **Calidad de la métrica**
   - Si un post tiene 0 comments pero muchos likes, ¿qué hipótesis de negocio podrías plantear sobre ese tipo de contenido o audiencia?  
   - ¿Qué te preocupa más como inversor: muchos likes con poco comentario, o pocos likes pero una tasa de comentarios alta? Explica.


```python
#1 Datos faltantes
columnas_clave = ["commentsCount", "viewsCount"]
faltantes = df[columnas_clave].isna().any(axis=1).sum()
total_posts = df.shape[0]
porcentaje_faltantes = (faltantes / total_posts) * 100

print(f"Porcentaje de posts con valores faltantes en métricas clave: {porcentaje_faltantes:.2f}%")

print("-----------------------------")

#2 Outliers
# Definir umbral para outliers (por ejemplo, 95 percentil)
umbral_comments = df['commentsCount'].quantile(0.95)
umbral_views = df['viewsCount'].quantile(0.95)

# Filtrar posts que superan esos umbrales
outliers = df[(df['commentsCount'] > umbral_comments) | (df['viewsCount'] > umbral_views)]

outliers[['brand', 'commentsCount', 'viewsCount', 'publishedAt']]

```

### **Respuestas**

1. Datos faltantes

* Desde el punto de vista de negocio, si un post tiene métricas faltantes, no conviene excluirlo automáticamente porque podríamos perder información valiosa sobre la marca. Tampoco siempre es recomendable imputar valores porque podríamos distorsionar la realidad de la interacción real. Lo más prudente sería analizar esos posts aparte: revisar por qué faltan los datos, identificar patrones (por ejemplo, si ciertos tipos de publicaciones o formatos suelen carecer de métricas) y decidir caso por caso cómo incluirlos en el análisis, dejando claros los límites de la información disponible.

2. Outliers

* Sobre la segunda pregunta: esos posts con números extremadamente altos podrían no ser completamente orgánicos; podrían deberse a campañas pagas, promociones especiales o contenido viral que alcanzó mucha visibilidad. Identificar esto es clave porque pueden distorsionar la percepción de la interacción promedio de la marca.
En cuanto a la tercera pregunta: si incluyes estos outliers en el análisis, los promedios de likes, comentarios o vistas se elevan, haciendo que la marca parezca más efectiva de lo que es de manera general. Si los excluyes, obtienes una lectura más representativa del desempeño típico, pero corres el riesgo de ignorar casos exitosos importantes que podrían reflejar la efectividad de campañas puntuales. Lo ideal es reportar ambos enfoques para tener una visión completa: promedio general y comportamiento típico sin outliers.

3. Calidad de la métrica

* Si un post tiene 0 comentarios pero muchos likes, podría indicar que el contenido es atractivo visualmente o fácil de consumir, pero no genera conversación ni discusión entre la audiencia. Tal vez sea contenido rápido, informativo o promocional que la gente aprueba con un like sin sentir la necesidad de interactuar más. También podría reflejar una audiencia más pasiva que prefiere reaccionar con likes en lugar de escribir comentarios.
Como inversor, me preocuparía más si veo muchos likes con pocos comentarios porque esto puede dar una falsa sensación de engagement profundo; el contenido parece popular, pero no está generando interacción real ni lealtad de la comunidad. Por otro lado, pocos likes pero una tasa de comentarios alta podría indicar una audiencia muy comprometida y activa, lo cual es valioso para construir relaciones a largo plazo y efectividad de campañas, aunque el alcance inicial sea menor.
